##### Going to test if wines from one country is significantly higher than other by first checking their normality. 

```python
us = wine11[wine11.country=="US"]
fr = wine11[wine11.country=="France"]
it = wine11[wine11.country=="Italy"]
ge = wine11[wine11.country=="Germany"]
ca = wine11[wine11.country=="Canada"]
au = wine11[wine11.country=='Australia']
six_countries = pd.concat([us, fr, it, ge, ca, au])

sns.set(rc={'figure.figsize':(10,8)})

fig, axes = plt.subplots(nrows=3, ncols=2)
fig.subplots_adjust(hspace=1)
fig.suptitle('Normality test for wine from each country')

plt.subplot(3,2,1)
us_plot = sns.distplot(us.price);
plt.title("US");
plt.text(400, 0.015,"P-value for nomality test: " + str('{:f}'.format(stats.normaltest(us.price)[1])));

plt.subplot(3,2,2)
fr_plot = sns.distplot(fr.price);
plt.title("France");
plt.text(500, 0.015,"P-value for nomality test: " + str('{:f}'.format(stats.normaltest(fr.price)[1])));

plt.subplot(3,2,3)
it_plot = sns.distplot(it.price);
plt.title("Italy");
plt.text(200, 0.023,"P-value for nomality test: " + str('{:f}'.format(stats.normaltest(it.price)[1])));

plt.subplot(3,2,4)
ge_plot = sns.distplot(ge.price);
plt.title("Germany");
plt.text(150, 0.026,"P-value for nomality test: " + str('{:f}'.format(stats.normaltest(ge.price)[1])));

plt.subplot(3,2,5)
ca_plot = sns.distplot(ca.price);
plt.title("Canada");
plt.text(31, 0.031,"P-value for nomality test: " + str('{:f}'.format(stats.normaltest(ca.price)[1])));

plt.subplot(3,2,6)
au_plot = sns.distplot(au.price);
plt.title("Australia");
plt.text(170, 0.031,"P-value for nomality test: " + str('{:f}'.format(stats.normaltest(au.price)[1])));
```

##### Since none of the data from these countries are normally distributed, I am going to define functions to perform Mann-whitney U test for not normally distributed dataset.

```python
def hypo_1side(x, y):
    """H0: German wine price is >= Italian wine price
         H1: German wine price is < Italian wine price"""
    if (stats.mannwhitneyu(x, y, alternative='less')[1]) >= 0.05:
        return "P-value: " + str('{:f}'.format(stats.mannwhitneyu(x, y, alternative='less')[1])) + ", we cannot reject the null hypothesis"
    else:
        return "P-value: " + str('{:f}'.format(stats.mannwhitneyu(x, y, alternative='less')[1])) + ", we can reject the null hypothesis, and accept the alternative"

def hypo_2sides(x, y):
    """H0: German wine price is == Italian wine price
         H1: German wine price is =! Italian wine price"""
    if (stats.mannwhitneyu(x, y, alternative='less')[1]) >= 0.05:
        return "P-value: " + str('{:f}'.format(stats.mannwhitneyu(x, y, alternative='two-sided')[1])) + ", we cannot reject the null hypothesis"
    else:
        return "P-value: " + str('{:f}'.format(stats.mannwhitneyu(x, y, alternative='two-sided')[1])) + ", we can reject the null hypothesis, and accept the alternative"
```

##### By putting in different variables into functions, we get the result of: Australian wine price is cheaper than French wine price < German wine price <Italian wine price == US wine price == Canadian wine price

```python
# # For countries that can reject
print(hypo_1side(au.price, ge.price)) # Australian wine price < German wine price
# print(hypo_1side(au.price, fr.price)) # Australian wine price < French wine price
# print(hypo_1side(fr.price, it.price)) # French wine price < Italian wine price
# print(hypo_1side(fr.price, ge.price)) # French wine price < German wine price
# print(hypo_1side(ge.price, ca.price)) # German wine price < Canadian wine price
# print(hypo_1side(ge.price, us.price)) # German wine price < US wine price
# print(hypo_1side(ge.price, it.price)) # German wine price < Italian wine price

print(hypo_1side(it.price, us.price)) # US wine price >= Italian wine price
# print(hypo_1side(ca.price, it.price)) # Canadian wine price == Italian wine price 
# print(hypo_1side(ca.price, us.price)) 
# print(hypo_1side(us.price, it.price)) # why?
# print(hypo_1side(it.price, ca.price))
# print(hypo_1side(us.price, ca.price))

# # For countries that can't reject both way(>= , <)
print(hypo_2sides(it.price, us.price)) # why # US wine price == Italian wine price
# print(hypo_2sides(ca.price, it.price)) # Canadian wine price == Italian wine price 
# print(hypo_2sides(ca.price, us.price)) # Canadian wine price == US wine price

# ## (power of the test for it and us)
```
